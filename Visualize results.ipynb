```python
import pandas as pd
```

```python
df = pd.read_csv('Classification Results 2.csv')
```

```python
df.sample(n=5)
```

```python
best_cases = pd.DataFrame(columns=df.columns)
for dataset in df.dataset.unique():
    # Za svaki classifier trazimo optimalne hiperparametre
    for classifier in df.classifier.unique():
        temp_df = df[(df['dataset'] == dataset) & (df['classifier'] == classifier)].copy()
        best_cases = best_cases.append(temp_df[temp_df['metric']=='recall'].max(), ignore_index=True)
```

```python
results = pd.DataFrame(columns = df.columns)
for i, row in best_cases.iterrows():
    results = results.append(
        df[(df.dataset == row.dataset) & (df.classifier == row.classifier) & (df.hyperparameter == row.hyperparameter)],
        ignore_index=True
    )
```

```python
results
```

```python
results.to_csv('Best Hyperparameters.csv', index=False)
```
