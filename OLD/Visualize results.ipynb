```python
import pandas as pd
```

```python
df = pd.read_csv('Classification Results 2.csv')
```

```python
df.sample(n=5)
```

```python
best_cases = pd.DataFrame(columns=df.columns)
for dataset in df.dataset.unique():
    # Za svaki classifier trazimo optimalne hiperparametre
    for classifier in df.classifier.unique():
        temp_df = df[(df['dataset'] == dataset) & (df['classifier'] == classifier)].copy()
        best_cases = best_cases.append(temp_df[temp_df['metric']=='recall'].max(), ignore_index=True)
```

```python
results = pd.DataFrame(columns = df.columns)
for i, row in best_cases.iterrows():
    results = results.append(
        df[(df.dataset == row.dataset) & (df.classifier == row.classifier) & (df.hyperparameter == row.hyperparameter)],
        ignore_index=True
    )
```

```python
results
```

```python
results.to_csv('Best Hyperparameters.csv', index=False)
```

```python
import seaborn as sns

for dataset in results.dataset.unique():
    # Draw a nested barplot by species and sex
    g = sns.catplot(
        data=results[results['dataset'] == dataset], kind="bar",
        x="classifier", y="value", hue="metric",
        ci="sd", palette='bright', alpha=.8, height=6
    )
    g.despine(left=True)
    g.set_axis_labels("Metod", "")
    g.legend.set_title(dataset)
    g.savefig('{}.png'.format(dataset))
```

```python
import matplotlib.pyplot as plt

for classifier in results.classifier.unique():
    # Draw a nested barplot by species and sex
    g = sns.catplot(
        data=results[results.classifier == classifier], kind="bar",
        x="dataset", y="value", hue="metric",
        ci="sd", palette="bright", alpha=.8, height=6
    )
    g.despine(left=True)
    g.set_axis_labels("Skup podataka", "")
    g.legend.set_title(classifier)
    plt.xticks(rotation=45)
    g.savefig('{}.png'.format(classifier))
```
